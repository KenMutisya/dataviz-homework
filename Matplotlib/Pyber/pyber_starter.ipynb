```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
merged_data = pd.merge(ride_data, city_data, how='left')

# Display the data table for preview
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x1 = merged_data.loc[merged_data["type"] == "Urban", :]
x1 = x1.groupby(['city', 'type']).count()['fare']
x1_df = pd.DataFrame({'Total Number of Rides': x1})

x2 = merged_data.loc[merged_data["type"] == "Suburban", :]
x2 = x2.groupby(['city', 'type']).count()['fare']
x2_df = pd.DataFrame({'Total Number of Rides': x2})

x3 = merged_data.loc[merged_data["type"] == "Rural", :]
x3 = x3.groupby(['city', 'type']).count()['fare']
x3_df = pd.DataFrame({'Total Number of Rides': x3})

y1 = merged_data.loc[merged_data["type"] == "Urban", :]
y1 = y1.groupby(['city', 'type']).mean()['fare']
y1_df = pd.DataFrame({'Average Fare': y1})

y2 = merged_data.loc[merged_data["type"] == "Suburban", :]
y2 = y2.groupby(['city', 'type']).mean()['fare']
y2_df = pd.DataFrame({'Average Fare': y2})

y3 = merged_data.loc[merged_data["type"] == "Rural", :]
y3 = y3.groupby(['city', 'type']).mean()['fare']
y3_df = pd.DataFrame({'Average Fare': y3})

# City driver count data
driver_count = merged_data[["city", "driver_count", "type"]]
driver_count_urban = driver_count.loc[driver_count["type"] == "Urban", :]
driver_group_urban = driver_count_urban.groupby(["city"])
driver_comparison_urban = driver_group_urban.mean()

driver_count_suburban = driver_count.loc[driver_count["type"] == "Suburban", :]
driver_group_suburban = driver_count_suburban.groupby(["city"])
driver_comparison_suburban = driver_group_suburban.mean()

driver_count_rural = driver_count.loc[driver_count["type"] == "Rural", :]
driver_group_rural = driver_count_rural.groupby(["city"])
driver_comparison_rural = driver_group_rural.mean()

# Build the scatter plots for each city types
plt.scatter(x1, y1, marker="o", facecolors="lightcoral", edgecolors="black", s=(driver_comparison_urban)*8, alpha=0.8, linewidth=1.25, label="Urban")
plt.scatter(x2, y2, marker="o", facecolors="lightskyblue", edgecolors="black", s=(driver_comparison_suburban)*8, alpha=0.8, linewidth=1.25, label="Suburban")
plt.scatter(x3, y3, marker="o", facecolors="gold", edgecolors="black", s=(driver_comparison_rural)*8, alpha=0.8, linewidth=1.25, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
plt.legend(loc="best", title="City Types")

# Incorporate a text label regarding circle size
plt.text(42, 38, "Note: Circle size correlates with driver count per city.")

# Save Figure
plt.savefig("Images/BubblePlot.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fares = merged_data.loc[merged_data["type"] == "Urban", :]
urban_fares = urban_fares[['fare']].sum()

suburban_fares = merged_data.loc[merged_data["type"] == "Suburban", :]
suburban_fares = suburban_fares[['fare']].sum()

rural_fares = merged_data.loc[merged_data["type"] == "Rural", :]
rural_fares = rural_fares[['fare']].sum()

# Build Pie Chart
city_types = ["Urban", "Rural", "Suburban"]
fare_data = [urban_fares, rural_fares, suburban_fares]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1,0,0)
plt.axis("equal")

plt.pie(fare_data, explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=300)

plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("Images/TotalFares.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_rides = merged_data.loc[merged_data["type"] == "Urban", :]
urban_rides = urban_rides[['fare']].count()

suburban_rides = merged_data.loc[merged_data["type"] == "Suburban", :]
suburban_rides = suburban_rides[['fare']].count()

rural_rides = merged_data.loc[merged_data["type"] == "Rural", :]
rural_rides = rural_rides[['fare']].count()

# Build Pie Chart
city_types = ["Urban", "Rural", "Suburban"]
rides_data = [urban_rides, rural_rides, suburban_rides]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1,0,0)
plt.axis("equal")

plt.pie(rides_data, explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=275)

plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Images/TotalRides.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_drivers = city_data.loc[city_data["type"] == "Urban", "driver_count"]
urban_drivers = urban_drivers.sum()

suburban_drivers = city_data.loc[city_data["type"] == "Suburban", "driver_count"]
suburban_drivers = suburban_drivers.sum()

rural_drivers = city_data.loc[city_data["type"] == "Rural", "driver_count"]
rural_drivers = rural_drivers.sum()

# Build Pie Chart
city_types = ["Urban", "Rural", "Suburban"]
drivers_data = [urban_drivers, rural_drivers, suburban_drivers]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1,0,0)
plt.axis("equal")

plt.pie(drivers_data, explode=explode, labels=city_types, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=225)

plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Images/TotalDrivers.png")

# Show Figure
plt.show()
```
