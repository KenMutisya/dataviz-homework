```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
merged_data = pd.merge(city_data, ride_data, how='outer', on='city')

# Display the data table for preview
```

```python
merged_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x1 = merged_data.loc[merged_data["type"] == "Urban", :]
x1 = x1.groupby(['city', 'type']).count()['fare']
x1_df = pd.DataFrame({'Total Number of Rides': x1})

x2 = merged_data.loc[merged_data["type"] == "Suburban", :]
x2 = x2.groupby(['city', 'type']).count()['fare']
x2_df = pd.DataFrame({'Total Number of Rides': x2})

x3 = merged_data.loc[merged_data["type"] == "Rural", :]
x3 = x3.groupby(['city', 'type']).count()['fare']
x3_df = pd.DataFrame({'Total Number of Rides': x3})

y1 = merged_data.loc[merged_data["type"] == "Urban", :]
y1 = y1.groupby(['city', 'type']).mean()['fare']
y1_df = pd.DataFrame({'Average Fare': y1})

y2 = merged_data.loc[merged_data["type"] == "Suburban", :]
y2 = y2.groupby(['city', 'type']).mean()['fare']
y2_df = pd.DataFrame({'Average Fare': y2})

y3 = merged_data.loc[merged_data["type"] == "Rural", :]
y3 = y3.groupby(['city', 'type']).mean()['fare']
y3_df = pd.DataFrame({'Average Fare': y3})

# City driver count data
driver_count = merged_data[["city", "driver_count", "type"]]
driver_count_urban = driver_count.loc[driver_count["type"] == "Urban", :]
driver_group_urban = driver_count_urban.groupby(["city"])
driver_comparison_urban = driver_group_urban.mean()

driver_count_suburban = driver_count.loc[driver_count["type"] == "Suburban", :]
driver_group_suburban = driver_count_suburban.groupby(["city"])
driver_comparison_suburban = driver_group_suburban.mean()

driver_count_rural = driver_count.loc[driver_count["type"] == "Rural", :]
driver_group_rural = driver_count_rural.groupby(["city"])
driver_comparison_rural = driver_group_rural.mean()

# Build the scatter plots for each city types
plt.scatter(x1, y1, marker="o", facecolors="red", edgecolors="black", s=(driver_comparison_urban)*8, alpha=0.5, label="Urban")
plt.scatter(x2, y2, marker="o", facecolors="blue", edgecolors="black", s=(driver_comparison_suburban)*8, alpha=0.5, label="Suburban")
plt.scatter(x3, y3, marker="o", facecolors="yellow", edgecolors="black", s=(driver_comparison_rural)*8, alpha=0.5, label="Rural")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
plt.legend(loc="best", title="City Types")

# Incorporate a text label regarding circle size
plt.text(42, 38, "Note: Circle size correlates with driver count per city.")

# Save Figure
```

```python
# Show plot
plt.show()
```
